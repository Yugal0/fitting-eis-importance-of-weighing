The purpose of this is to sort the 3D aspect ratio problem in matplotlib, bcoz ineed that in tommorows presentation

```python
%matplotlib ipympl
import numpy as np
import scipy
import matplotlib.pyplot as plt

fig=plt.figure()
ax=fig.add_subplot(projection="3d")
x=np.linspace(0,10,100)
r=10
def give_y(x):
    return np.sqrt(r**2-x**2)
ax.plot(x,give_y(x),np.zeros(len(x)))
ax.set_box_aspect([1,1,1])
```
